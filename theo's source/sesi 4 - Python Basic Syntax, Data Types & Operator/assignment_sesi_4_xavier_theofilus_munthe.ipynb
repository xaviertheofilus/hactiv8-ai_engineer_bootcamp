```python
produk = [

    {"nama": "Laptop ASUS A516", "harga": 7500000, "stok": 12,

     "kategori": "laptop"},

    {"nama": "Mouse Logitech M100", "harga": 85000, "stok": 50,

     "kategori": "aksesoris"},

    {"nama": "Keyboard Mechanical Rexus", "harga": 450000, "stok": 30,

     "kategori": "aksesoris"},

    {"nama": "Monitor LG 24 inch", "harga": 2300000, "stok": 8,

     "kategori": "monitor"},

    {"nama": "SSD Kingston 512GB", "harga": 650000, "stok": 25,

     "kategori": "storage"},

]
```

```python
# Task 1

for p in produk:
    kategori = p["kategori"].upper()
    nama = p["nama"]
    harga = f"{p['harga']:,}".replace(",", ".")
    stok = p["stok"]

    print(f"[{kategori}] {nama} | Harga: Rp {harga} | Stok: {stok} unit")
```

```python
# Task 1

print("-" * 80)
print(f"{'KATEGORI':<12} | {'NAMA PRODUK':<30} | {'HARGA':>14} | {'STOK':>5}")
print("-" * 80)

for p in produk:
    kategori = p["kategori"].upper()
    nama = p["nama"]
    harga = f"{p['harga']:,}".replace(",", ".")
    stok = p["stok"]

    print(f"{kategori:<12} | {nama:<30} | Rp {harga:>11} | {stok:>4}")
```

```python
# Task 2

total_nilai_inventory = 0

for p in produk:
    total_nilai_inventory += p["harga"] * p["stok"]

print(f"Total Nilai Inventory: Rp {total_nilai_inventory:,}".replace(",", "."))

rata_harga = total_nilai_inventory / sum(p["stok"] for p in produk)
print(f"Rata-rata Harga Produk: Rp {rata_harga:.2f}")
```

```python
# Task 3

nama_produk = " | ".join([p["nama"] for p in produk])

print("Daftar Nama Produk:")
print(nama_produk)

jumlah_karakter = len(nama_produk)
print("Jumlah karakter:", jumlah_karakter)

print("Apakah mengandung kata 'Rexus':", "Rexus" in nama_produk)

nama_produk_pertama = nama_produk.split(" | ")[0]
print("Nama produk pertama:", nama_produk_pertama)
```

```python
# Task 4 - Cell 1

ringkasan_kategori = {}

for p in produk:
    kategori = p["kategori"]
    ringkasan_kategori[kategori] = ringkasan_kategori.get(kategori, 0) + 1

print("Ringkasan sebelum penambahan produk:")
for k, v in ringkasan_kategori.items():
    print(f"Kategori [{k}]: {v} produk")
```

```python
# lock data
produk_awal = produk[:5]

ringkasan_kategori = {}

for p in produk_awal:
    kategori = p["kategori"]
    ringkasan_kategori[kategori] = ringkasan_kategori.get(kategori, 0) + 1

print("Ringkasan sebelum penambahan produk:")
for k, v in ringkasan_kategori.items():
    print(f"Kategori [{k}]: {v} produk")
```

```python
# Task 4 - Cell 2
nama_produk_baru = "Webcam Logitech C920"
sudah_ada = False

for p in produk:
    if p["nama"] == nama_produk_baru:
        sudah_ada = True
        break

if not sudah_ada:
    produk.append({
        "nama": "Webcam Logitech C920",
        "harga": 750000,
        "stok": 15,
        "kategori": "aksesoris"
    })

ringkasan_kategori = {}

for p in produk:
    kategori = p["kategori"]
    ringkasan_kategori[kategori] = ringkasan_kategori.get(kategori, 0) + 1

print("\nRingkasan setelah penambahan produk:")
for k, v in ringkasan_kategori.items():
    print(f"Kategori [{k}]: {v} produk")
```
