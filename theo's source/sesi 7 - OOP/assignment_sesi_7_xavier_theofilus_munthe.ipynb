```python
# Task 1 Base class Karyawan

class Karyawan:

    def __init__(self, nama, departemen, gaji_pokok):
        self.nama = nama
        self.departemen = departemen
        self.__gaji_pokok = gaji_pokok

    def get_gaji_pokok(self):
        return self.__gaji_pokok

    def set_gaji_pokok(self, nilai):
        if nilai > 0:
            self.__gaji_pokok = nilai
        else:
            print("Error: gaji pokok harus lebih dari 0")

    def hitung_gaji(self):
        return self.__gaji_pokok

    def info(self):
        print(f"Nama       : {self.nama}")
        print(f"Departemen : {self.departemen}")
        print(f"Gaji Pokok : Rp {self.get_gaji_pokok():,}")
        print(f"Total Gaji : Rp {self.hitung_gaji():,}")
```

```python
# test task 1
x = Karyawan("theo", "IT", 5000000)
x.info()
```

```python
# Task 2 Subclass KaryawanTetap

class KaryawanTetap(Karyawan):

    def __init__(self, nama, departemen, gaji_pokok, tunjangan_tetap, bonus_tahunan):
        super().__init__(nama, departemen, gaji_pokok)
        self.tunjangan_tetap = tunjangan_tetap
        self.bonus_tahunan = bonus_tahunan

    def hitung_gaji(self):
        return int(self.get_gaji_pokok() + self.tunjangan_tetap + (self.bonus_tahunan / 12))

    def info(self):
        print(f"Jenis      : Karyawan Tetap")
        print(f"Nama       : {self.nama}")
        print(f"Departemen : {self.departemen}")
        print(f"Gaji Pokok : Rp {self.get_gaji_pokok():,}")
        print(f"Tunjangan  : Rp {self.tunjangan_tetap:,}")
        print(f"Bonus/bln  : Rp {int(self.bonus_tahunan / 12):,}")
        print(f"Total Gaji : Rp {self.hitung_gaji():,}")


kt1 = KaryawanTetap("alek", "Engineering", 8000000, 1500000, 12000000)
kt2 = KaryawanTetap("bobe", "Product", 7500000, 1200000, 9600000)

kt1.info()
print()
kt2.info()
```

```python
# Task 3 Subclass KaryawanKontrak

class KaryawanKontrak(Karyawan):

    def __init__(self, nama, departemen, gaji_pokok, durasi_kontrak_bulan, rate_per_jam, jam_per_bulan):
        super().__init__(nama, departemen, gaji_pokok)
        self.durasi_kontrak_bulan = durasi_kontrak_bulan
        self.rate_per_jam = rate_per_jam
        self.jam_per_bulan = jam_per_bulan

    def hitung_gaji(self):
        return self.rate_per_jam * self.jam_per_bulan

    def info(self):
        print(f"Jenis      : Karyawan Kontrak")
        print(f"Nama       : {self.nama}")
        print(f"Departemen : {self.departemen}")
        print(f"Durasi     : {self.durasi_kontrak_bulan} bulan")
        print(f"Rate/jam   : Rp {self.rate_per_jam:,}")
        print(f"Jam/bulan  : {self.jam_per_bulan} jam")
        print(f"Total Gaji : Rp {self.hitung_gaji():,}")


kk1 = KaryawanKontrak("zano", "IT Support", 0, 6, 75000, 160)
kk2 = KaryawanKontrak("salomo", "QA Engineer", 0, 3, 65000, 120)

kk1.info()
print()
kk2.info()
```

```python
# Task 4 Polymorphism dan manajemen list

tim_techspark = [kt1, kt2, kk1, kk2]

# polymorphism:
print("=== Daftar Gaji Tim Techspark ===")
total_gaji = 0
for karyawan in tim_techspark:
    gaji = karyawan.hitung_gaji()
    print(f"{karyawan.nama} ({type(karyawan).__name__}): Rp {gaji:,}")
    total_gaji += gaji

print(f"\nTotal pengeluaran gaji/bulan: Rp {total_gaji:,}")

print("\n=== Sebelum kenaikan gaji ===")
kt1.info()

kt1.set_gaji_pokok(kt1.get_gaji_pokok() + 500000)

print("\n=== Sesudah kenaikan gaji ===")
kt1.info()

try:
    print(kt1.__gaji_pokok)
except AttributeError as e:
    print(f"\nAttributeError terbukti: {e}")
```
