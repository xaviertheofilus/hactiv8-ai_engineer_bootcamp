Set the matching environment variable (`AZURE_OPENAI_*`, `GEMINI_API_KEY`, or `GROQ_API_KEY`) before running the cell.

> âš ï¸ **Groq does not provide embedding models** â€” Section 2.2 / 2.3 (embedding cells) will fail on Groq. Use Azure or Gemini if you want to run the full notebook.

```python
# Install dependencies (run once per environment).
# The `openai` SDK works for Azure OpenAI, Gemini, and Groq â€” they all
# expose OpenAI-compatible endpoints, so one SDK covers all three.
%pip install -q openai tiktoken python-dotenv numpy

import os
from dotenv import load_dotenv
load_dotenv()  # reads .env file into os.environ (placed alongside this notebook)
import time
import json
import numpy as np
from typing import List, Dict

# ============================================================
# CHOOSE YOUR PROVIDER: "azure", "gemini", or "groq"
# ============================================================
PROVIDER = "azure"   # change to "gemini" or "groq" to switch

if PROVIDER == "azure":
    from openai import AzureOpenAI
    client = AzureOpenAI(
        azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT", "https://YOUR-RESOURCE.openai.azure.com/"),
        api_key=os.getenv("AZURE_OPENAI_API_KEY", "YOUR-API-KEY"),
        api_version="2024-12-01-preview",
    )
    CHAT_MODEL       = "gpt-4o"
    CHAT_MODEL_MINI  = "gpt-4o-mini"
    EMBEDDING_MODEL  = "text-embedding-3-small"

elif PROVIDER == "gemini":
    # Google AI Studio â€” get a free key at https://aistudio.google.com/app/apikey
    from openai import OpenAI
    client = OpenAI(
        api_key=os.getenv("GEMINI_API_KEY", "YOUR-GEMINI-API-KEY"),
        base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    )
    CHAT_MODEL       = "gemini-2.0-flash"
    CHAT_MODEL_MINI  = "gemini-2.0-flash-lite"
    EMBEDDING_MODEL  = "text-embedding-004"

elif PROVIDER == "groq":
    # Groq Cloud â€” get a free key at https://console.groq.com/keys
    from openai import OpenAI
    client = OpenAI(
        api_key=os.getenv("GROQ_API_KEY", "YOUR-GROQ-API-KEY"),
        base_url="https://api.groq.com/openai/v1",
    )
    CHAT_MODEL       = "llama-3.3-70b-versatile"
    CHAT_MODEL_MINI  = "llama-3.1-8b-instant"
    EMBEDDING_MODEL  = None   # Groq does not provide embedding models

else:
    raise ValueError(f"Unknown PROVIDER: {PROVIDER!r}. Use 'azure', 'gemini', or 'groq'.")


# â”€â”€ Sanity check: catch the most common mistake (no API key set) â”€â”€
def _check_key(name, value, placeholder):
    if value == placeholder or not value:
        raise RuntimeError(
            f"âŒ {name} is not set. Either export it in your shell or put it in a `.env` "
            f"file next to this notebook (then re-run this cell). "
            f"Current value looks like the placeholder."
        )
if PROVIDER == "azure":
    _check_key("AZURE_OPENAI_API_KEY", os.getenv("AZURE_OPENAI_API_KEY", ""), "YOUR-API-KEY")
elif PROVIDER == "gemini":
    _check_key("GEMINI_API_KEY", os.getenv("GEMINI_API_KEY", ""), "YOUR-GEMINI-API-KEY")
elif PROVIDER == "groq":
    _check_key("GROQ_API_KEY", os.getenv("GROQ_API_KEY", ""), "YOUR-GROQ-API-KEY")

print(f"âœ… {PROVIDER.upper()} client initialized")
print(f"   Chat model:      {CHAT_MODEL}")
print(f"   Mini model:      {CHAT_MODEL_MINI}")
print(f"   Embedding model: {EMBEDDING_MODEL or '(not supported on this provider)'}")
```

```python
# ============================================================
# HELPER FUNCTIONS
# ============================================================

def chat_completion(messages: List[Dict], model: str = None,
                    temperature: float = 0.0, max_tokens: int = 500) -> str:
    """Call chat completion (works on Azure / Gemini / Groq)."""
    response = client.chat.completions.create(
        model=model or CHAT_MODEL,
        messages=messages,
        temperature=temperature,
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content


def get_embedding(text: str) -> List[float]:
    """Get embedding vector. Not supported on Groq."""
    if EMBEDDING_MODEL is None:
        raise RuntimeError(
            f"Embeddings are not supported on provider {PROVIDER!r}. "
            "Switch PROVIDER to 'azure' or 'gemini' to run embedding cells."
        )
    response = client.embeddings.create(input=text, model=EMBEDDING_MODEL)
    return response.data[0].embedding


def cosine_similarity(a: List[float], b: List[float]) -> float:
    """Compute cosine similarity between two vectors."""
    a, b = np.array(a), np.array(b)
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))

print("âœ… Helper functions loaded")
```

## 2. NLP Foundations

### 2.1 Tokenization

Tokenization is the very first step â€” breaking text into smaller units (tokens) and mapping them to numbers. LLMs don't see words; they see token IDs.

```python
# ============================================================
# 2.1 TOKENIZATION â€” How machines read our text
# ============================================================
import tiktoken

# Load the tokenizer used by GPT-4o
encoder = tiktoken.encoding_for_model("gpt-4o")

# â”€â”€ Example 1: Simple English sentence â”€â”€
text_en = "I love my cat"
tokens_en = encoder.encode(text_en)

print("â”€â”€ English Tokenization â”€â”€")
print(f"  Text:    '{text_en}'")
print(f"  Tokens:  {tokens_en}")
print(f"  Count:   {len(tokens_en)} tokens")
print(f"  Decoded: {[encoder.decode([t]) for t in tokens_en]}")

# â”€â”€ Example 2: Bahasa Indonesia (uses more tokens!) â”€â”€
text_id = "Saya suka kucing saya"
tokens_id = encoder.encode(text_id)

print(f"\nâ”€â”€ Bahasa Indonesia Tokenization â”€â”€")
print(f"  Text:    '{text_id}'")
print(f"  Tokens:  {tokens_id}")
print(f"  Count:   {len(tokens_id)} tokens")
print(f"  Decoded: {[encoder.decode([t]) for t in tokens_id]}")

# â”€â”€ Example 3: Code snippet â”€â”€
text_code = "def hello():\n    print('Hello, world!')"
tokens_code = encoder.encode(text_code)

print(f"\nâ”€â”€ Code Tokenization â”€â”€")
print(f"  Text:    '{text_code}'")
print(f"  Count:   {len(tokens_code)} tokens")

# â”€â”€ Key insight: same meaning, different token counts â”€â”€
print(f"\nðŸ’¡ Key Insight:")
print(f"   English '{text_en}' = {len(tokens_en)} tokens")
print(f"   Indonesian '{text_id}' = {len(tokens_id)} tokens")
print(f"   Indonesian uses ~{len(tokens_id)/len(tokens_en):.1f}x more tokens for similar meaning!")
```

### 2.2 Embeddings

Embeddings map text into a vector space where **similar meanings** are placed **closer together**. This is the foundation of semantic search and RAG.

> âš ï¸ **Groq users:** the next two cells call `get_embedding()` and will raise an error. Switch `PROVIDER` to `"azure"` or `"gemini"` in the setup cell to run these.

```python
# ============================================================
# 2.2 EMBEDDINGS â€” Turning words into numbers
# ============================================================

# Embed several words and compare their similarity
words = ["king", "queen", "prince", "apple", "banana", "fruit"]
embeddings = {w: get_embedding(w) for w in words}

print("â”€â”€ Embedding Properties â”€â”€")
print(f"  Dimensions: {len(embeddings['king'])}")
print(f"  First 5 values of 'king': {embeddings['king'][:5]}")

# â”€â”€ Semantic similarity matrix â”€â”€
print(f"\nâ”€â”€ Semantic Similarity Matrix â”€â”€")
print(f"{'':>10}", end="")
for w in words:
    print(f"{w:>10}", end="")
print()

for w1 in words:
    print(f"{w1:>10}", end="")
    for w2 in words:
        sim = cosine_similarity(embeddings[w1], embeddings[w2])
        print(f"{sim:>10.3f}", end="")
    print()

print(f"\nðŸ’¡ Notice: 'king' â†” 'queen' ({cosine_similarity(embeddings['king'], embeddings['queen']):.3f}) "
      f"is much higher than 'king' â†” 'apple' ({cosine_similarity(embeddings['king'], embeddings['apple']):.3f})")
```

```python
# ============================================================
# 2.3 SENTENCE EMBEDDINGS â€” Semantic search preview
# ============================================================

# Embed full sentences and find the most similar to a query
documents = [
    "The quarterly revenue grew by 15% driven by enterprise sales.",
    "Our new leave policy allows 20 days of paid time off per year.",
    "Cloud migration projects contributed to 45% of total revenue.",
    "The company hired 7 new engineers for the AI/ML team.",
    "Customer satisfaction scores reached an all-time high of 94%.",
]

query = "How did the company's income change?"
query_emb = get_embedding(query)

# Rank by similarity
scored = []
for i, doc in enumerate(documents):
    doc_emb = get_embedding(doc)
    sim = cosine_similarity(query_emb, doc_emb)
    scored.append((sim, i, doc))

scored.sort(reverse=True)

print(f"ðŸ”Ž Query: '{query}'")
print(f"\nâ”€â”€ Results ranked by semantic similarity â”€â”€")
for rank, (sim, idx, doc) in enumerate(scored):
    marker = "â­" if rank < 2 else "  "
    print(f"  {marker} [{sim:.4f}] {doc[:80]}...")

print(f"\nðŸ’¡ The model understands 'income change' â‰ˆ 'revenue grew' â€” this is semantic search!")
```

## 3. Inside the LLM: Text Generation

### 3.1 Autoregressive Generation

LLMs generate text **one token at a time**. Each new token is predicted based on all previous tokens. This is the autoregressive loop.

```python
# ============================================================
# 3.1 AUTOREGRESSIVE GENERATION â€” Watching token-by-token output
# ============================================================

messages = [
    {"role": "system", "content": "You are a helpful assistant. Keep responses brief."},
    {"role": "user", "content": "Explain what a Transformer is in 2 sentences."}
]

# Stream the response to see tokens arrive one at a time
print("â”€â”€ Streaming (token-by-token generation) â”€â”€")
print("ðŸ¤– ", end="")

stream = client.chat.completions.create(
    model=CHAT_MODEL,
    messages=messages,
    temperature=0.0,
    max_tokens=150,
    stream=True,  # Enable streaming
)

full_response = ""
for chunk in stream:
    if chunk.choices[0].delta.content:
        token = chunk.choices[0].delta.content
        full_response += token
        print(token, end="", flush=True)

print(f"\n\nðŸ“Š Total tokens generated: {len(encoder.encode(full_response))}")
```

### 3.2 Temperature & Sampling

**Temperature** controls randomness in token selection:
- `temperature=0` â†’ Always picks the most probable token (deterministic)
- `temperature=1` â†’ Samples proportional to probabilities (creative)
- `temperature>1` â†’ Even more random (chaotic)

```python
# ============================================================
# 3.2 TEMPERATURE â€” How LLMs pick the next token
# ============================================================

prompt = "Write a one-sentence tagline for a coffee shop called 'The Daily Grind'."

temperatures = [0.0, 0.5, 1.0, 1.5]

print("â”€â”€ Same prompt, different temperatures â”€â”€")
print(f"   Prompt: '{prompt}'\n")

for temp in temperatures:
    responses = []
    for _ in range(3):  # Generate 3 times to show variability
        messages = [{"role": "user", "content": prompt}]
        response = chat_completion(messages, temperature=temp, max_tokens=60)
        responses.append(response.strip())

    print(f"ðŸŒ¡ï¸  temperature={temp}:")
    for i, r in enumerate(responses):
        marker = "  âœ“" if i == 0 else "   "
        print(f"  {marker} {r}")

    # Check if all 3 are identical (deterministic)
    if len(set(responses)) == 1:
        print(f"  â†’ All identical (deterministic)")
    else:
        print(f"  â†’ {len(set(responses))} unique responses (creative)")
    print()
```

## 4. Azure OpenAI Models

### 4.1 Basic Chat Completion

The Azure OpenAI chat API uses three message roles: `system`, `user`, and `assistant`.

```python
# ============================================================
# 4.1 BASIC CHAT COMPLETION â€” The three message roles
# ============================================================

messages = [
    # SYSTEM: Sets persona, rules, and constraints (highest trust)
    {
        "role": "system",
        "content": (
            "You are a senior Azure Solutions Architect. "
            "Respond concisely in 2-3 sentences. "
            "Focus only on Azure services, not AWS or GCP."
        )
    },
    # USER: The actual question from the end-user
    {
        "role": "user",
        "content": "What's the best way to deploy a Python web app on Azure?"
    },
]

response = client.chat.completions.create(
    model=CHAT_MODEL,
    messages=messages,
    temperature=0.0,
    max_tokens=200,
)

result = response.choices[0]

print("â”€â”€ Chat Completion Result â”€â”€")
print(f"  Model:          {response.model}")
print(f"  Finish reason:  {result.finish_reason}")
print(f"  Input tokens:   {response.usage.prompt_tokens}")
print(f"  Output tokens:  {response.usage.completion_tokens}")
print(f"  Total tokens:   {response.usage.total_tokens}")
print(f"\nðŸ¤– Answer: {result.message.content}")
```

### 4.2 Multi-Turn Conversation

The model has **no memory**. You must resend the full message history every turn to maintain context.

```python
# ============================================================
# 4.2 MULTI-TURN CONVERSATION â€” State management
# ============================================================

# The model has NO memory â€” we reconstruct context by resending history
conversation = [
    {"role": "system", "content": "You are a helpful Azure support agent. Be concise."},
]

exchanges = [
    "How do I enable diagnostic logs on Azure App Service?",
    "Can I send them to a Log Analytics workspace?",
    "What about costs for that?",
]

print("â”€â”€ Multi-Turn Conversation â”€â”€\n")

for user_msg in exchanges:
    # Add user message to history
    conversation.append({"role": "user", "content": user_msg})

    # Call API with FULL history
    response = chat_completion(conversation, max_tokens=150)

    # Add assistant response to history
    conversation.append({"role": "assistant", "content": response})

    # Count total tokens being sent
    total_chars = sum(len(m["content"]) for m in conversation)

    print(f"ðŸ‘¤ User:      {user_msg}")
    print(f"ðŸ¤– Assistant: {response[:150]}...")
    print(f"   ðŸ“Š History: {len(conversation)} messages, ~{total_chars} chars")
    print()

print(f"âš ï¸  Notice: By turn 3, we're sending {len(conversation)} messages every API call!")
print(f"   This is why context window management matters.")
```

## 5. Context Window Concepts

The **context window** is the total number of tokens an LLM can process at one time â€” it's the model's working memory.

```python
# ============================================================
# 5.1 TOKEN COUNTING â€” Know your budget
# ============================================================
import tiktoken

encoder = tiktoken.encoding_for_model("gpt-4o")

# Count tokens for different inputs
examples = {
    "Short question": "What is Azure?",
    "Medium prompt": "You are a senior Azure Solutions Architect with 10+ years of experience. "
                     "Help the user design a scalable microservices architecture on Azure. "
                     "Focus on AKS, API Management, and Cosmos DB.",
    "Long context": "\n".join([
        f"Document {i}: This is paragraph {i} of the company financial report. "
        f"Revenue grew by {i*2}% in sector {i}." for i in range(50)
    ]),
}

print("â”€â”€ Token Counting â”€â”€")
for label, text in examples.items():
    tokens = encoder.encode(text)
    print(f"  {label:20s}: {len(tokens):>6,} tokens  ({len(text):>6,} chars)")

# Context window limits
print(f"\nâ”€â”€ Azure OpenAI Context Window Limits â”€â”€")
models = {
    "gpt-4o":      128_000,
    "gpt-4o-mini": 128_000,
    "gpt-4":        8_192,
    "gpt-35-turbo": 16_384,
}
for model, limit in models.items():
    print(f"  {model:20s}: {limit:>10,} tokens")
```

```python
# ============================================================
# 5.2 CONTEXT WINDOW MANAGEMENT â€” Truncation strategy
# ============================================================

def manage_context_window(
    conversation: List[Dict],
    max_tokens: int = 4000,
    strategy: str = "truncate_oldest"
) -> List[Dict]:
    """
    Keep conversation within token budget.

    Strategies:
    - truncate_oldest: Remove oldest user/assistant turns (keep system message)
    - summarize: Summarize old turns into a single message
    """

    def count_tokens(msgs):
        return sum(len(encoder.encode(m["content"])) for m in msgs)

    if strategy == "truncate_oldest":
        # Always keep the system message (first message)
        system_msg = [conversation[0]] if conversation[0]["role"] == "system" else []
        history = conversation[len(system_msg):]

        # Remove oldest turns until we're under budget
        while count_tokens(system_msg + history) > max_tokens and len(history) > 2:
            history = history[2:]  # Remove oldest user+assistant pair
            print(f"   âœ‚ï¸  Truncated oldest turn. Remaining: {len(history)} messages")

        return system_msg + history

    return conversation


# Demo: Build a long conversation then truncate
demo_conversation = [
    {"role": "system", "content": "You are a helpful assistant."},
]
for i in range(20):
    demo_conversation.append({"role": "user", "content": f"Tell me fact #{i+1} about Azure services and cloud computing."})
    demo_conversation.append({"role": "assistant", "content": f"Fact #{i+1}: Azure provides over {100+i*5} services across compute, storage, networking, and AI."})

total_before = sum(len(encoder.encode(m["content"])) for m in demo_conversation)

print(f"â”€â”€ Context Window Management â”€â”€")
print(f"  Before: {len(demo_conversation)} messages, {total_before} tokens")

managed = manage_context_window(demo_conversation, max_tokens=500)
total_after = sum(len(encoder.encode(m["content"])) for m in managed)

print(f"  After:  {len(managed)} messages, {total_after} tokens")
print(f"  Saved:  {total_before - total_after} tokens ({(1 - total_after/total_before)*100:.0f}% reduction)")
```

## 6. Choosing the Right Model

Compare models across the 4 dimensions: **Quality**, **Speed**, **Cost**, and **Constraints**.

```python
# ============================================================
# 6. MODEL COMPARISON â€” Quality, Speed, Cost
# ============================================================

test_prompt = [
    {"role": "system", "content": "You are a helpful assistant."},
    {"role": "user", "content":
     "Explain the CAP theorem in distributed systems. "
     "Give a concrete Azure example. Keep it under 100 words."}
]

models_to_test = [CHAT_MODEL, CHAT_MODEL_MINI]

print("â”€â”€ Model Comparison â”€â”€")
print(f"   Prompt: 'Explain CAP theorem with Azure example (<100 words)'\n")

for model_name in models_to_test:
    start = time.time()

    response = client.chat.completions.create(
        model=model_name,
        messages=test_prompt,
        temperature=0.0,
        max_tokens=200,
    )

    elapsed = time.time() - start
    answer = response.choices[0].message.content
    usage = response.usage

    print(f"ðŸ¤– Model: {model_name}")
    print(f"   â±ï¸  Latency:       {elapsed:.2f}s")
    print(f"   ðŸ“Š Input tokens:   {usage.prompt_tokens}")
    print(f"   ðŸ“Š Output tokens:  {usage.completion_tokens}")
    print(f"   ðŸ“ Answer: {answer[:200]}...")
    print()

print("ðŸ’¡ Key Tradeoffs:")
print("   GPT-4o: Higher quality reasoning, slightly slower, higher cost")
print("   GPT-4o-mini: Faster, cheaper, good enough for many tasks")
print("   â†’ Choose based on your quality/speed/cost requirements!")
```
