```python
# Task 1 — Hitung tagihan per pelanggan

menu = {
    "Espresso": 25000,
    "Latte": 35000,
    "Cappuccino": 38000,
    "Croissant": 28000,
    "Sandwich": 42000,
    "Cheesecake": 45000,
}

transaksi = [
    {"nama": "Budi", "pesanan": ["Latte", "Croissant", "Croissant"]},
    {"nama": "Sari", "pesanan": ["Espresso", "Cheesecake"]},
    {"nama": "Dian", "pesanan": ["Cappuccino", "Sandwich", "Latte", "Cheesecake"]},
    {"nama": "Reza", "pesanan": ["Espresso"]},
    {"nama": "Ayu",  "pesanan": ["Latte", "Latte", "Sandwich", "Croissant", "Cheesecake"]},
]

total_pendapatan = 0

for pelanggan in transaksi:
    nama = pelanggan["nama"]
    pesanan = pelanggan["pesanan"]

    subtotal = 0
    for item in pesanan:
        subtotal += menu[item]

    if subtotal < 50000:
        persen_diskon = 0
    elif subtotal < 100000:
        persen_diskon = 10
    else:
        persen_diskon = 15

    jumlah_diskon = subtotal * persen_diskon / 100
    total = subtotal - jumlah_diskon

    total_pendapatan += total

    print(f"Pelanggan: {nama}")
    print(f"Pesanan  : {', '.join(pesanan)}")
    print(f"Subtotal : Rp {subtotal:,.0f}".replace(",", "."))
    print(f"Diskon   : {persen_diskon}% (Rp {jumlah_diskon:,.0f})".replace(",", "."))
    print(f"Total    : Rp {total:,.0f}".replace(",", "."))
    print("-" * 40)

print(f"\nTotal pendapatan kafe: Rp {total_pendapatan:,.0f}".replace(",", "."))
```

```python
# Task 2 — Cari pelanggan dengan tagihan tertinggi

tagihan_pelanggan = {}

for pelanggan in transaksi:
    nama = pelanggan["nama"]
    subtotal = 0
    for item in pelanggan["pesanan"]:
        subtotal += menu[item]

    persen_diskon = 0
    if subtotal >= 100000:
        persen_diskon = 15
    elif subtotal >= 50000:
        persen_diskon = 10

    tagihan_pelanggan[nama] = subtotal - (subtotal * persen_diskon / 100)

nama_tertinggi = None
tagihan_tertinggi = 0

for nama, tagihan in tagihan_pelanggan.items():
    if tagihan > tagihan_tertinggi:
        tagihan_tertinggi = tagihan
        nama_tertinggi = nama

print(f"Pelanggan dengan tagihan tertinggi: {nama_tertinggi}")
print(f"Tagihan: Rp {tagihan_tertinggi:,.0f}".replace(",", "."))
```

```python
# Task 3 — Simulasi antrian dengan while loop

antrian = ["Budi", "Sari", "Dian", "Reza", "Ayu"]
stok_kopi = 3

item_kopi = ["Espresso", "Latte", "Cappuccino"]

while len(antrian) > 0 and stok_kopi > 0:
    pelanggan_sekarang = antrian.pop(0)

    # cari pesanan pelanggan dari data transaksi
    pesanan_pelanggan = []
    for p in transaksi:
        if p["nama"] == pelanggan_sekarang:
            pesanan_pelanggan = p["pesanan"]
            break

    jumlah_kopi = 0
    for item in pesanan_pelanggan:
        if item in item_kopi:
            jumlah_kopi += 1

    stok_kopi -= jumlah_kopi
    if stok_kopi < 0:
        stok_kopi = 0

    print(f"Melayani: {pelanggan_sekarang} | Stok kopi tersisa: {stok_kopi}")

if stok_kopi == 0 and len(antrian) > 0:
    print("Stok kopi habis. Pelanggan berikutnya tidak dapat dilayani.")
else:
    print("Semua pelanggan telah dilayani.")
```
