```python
import requests

# Task 1 — Ambil dan eksplorasi data dari public API

users_response = requests.get("https://jsonplaceholder.typicode.com/users")
posts_response = requests.get("https://jsonplaceholder.typicode.com/posts")

print("Status users:", users_response.status_code)
print("Status posts:", posts_response.status_code)

users = users_response.json()
posts = posts_response.json()

print(f"\nJumlah user  : {len(users)}")
print(f"Jumlah post  : {len(posts)}")

print("\n3 user pertama:")
for user in users[:3]:
    print(f"  {user['name']} — {user['email']}")

print("\n5 post pertama:")
for post in posts[:5]:
    print(f"  {post['title']}")

# hitung jumlah post per user
posts_per_user = {}
for post in posts:
    uid = post["userId"]
    if uid not in posts_per_user:
        posts_per_user[uid] = 0
    posts_per_user[uid] += 1

print("\nPosts per user:")
print(posts_per_user)
```

```python
# Task 2 — Proses dan transformasi data

user_summary = []
for user in users:
    user_summary.append({
        "id": user["id"],
        "nama": user["name"],
        "email": user["email"],
        "kota": user["address"]["city"],
        "jumlah_post": posts_per_user.get(user["id"], 0)
    })

user_summary = sorted(user_summary, key=lambda u: u["jumlah_post"], reverse=True)

print("\n3 user dengan post terbanyak:")
for i, user in enumerate(user_summary[:3], start=1):
    print(f"{i}. {user['nama']} ({user['kota']}) — {user['jumlah_post']} post")
```
